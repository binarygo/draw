```python
use_attn=True

face_cfg = draw_face.make_face_config(
    use_attn=use_attn,
    train_dir=("attn_" if use_attn else "") + "face_train_log")

tf.train.get_checkpoint_state(face_cfg.train_dir)
```

```python
gen_xs, gen_write_attn_infos = draw_util.eval_draw_model(face_cfg)
```

```python
batch_idx = 11
for i, gen_x in enumerate(gen_xs):
    x = gen_x[batch_idx,:,:]
    plt.imshow(x)
    plt.set_cmap("gray")
    if face_cfg.use_attn and i > 0:
        write_attn_info = [
            p[batch_idx] for p in gen_write_attn_infos[i - 1]
        ]
        gx, gy, sigma2, delta, gamma = write_attn_info
        w = delta * (face_cfg.write_width - 1.0)
        h = delta * (face_cfg.write_height - 1.0)
        print "debug: gx=%f, gy=%f, w=%f, h=%f, \nsigma2=%f, delta=%f, gamma=%f"%(
            gx, gy, w, h, sigma2, delta, gamma)
        g = (gx - w / 2.0, gy - h / 2.0)
        ax = plt.gca()
        ax.add_patch(patches.Rectangle(g, w, h, fill=False, edgecolor="r"))
    # plt.axis("off");    
    plt.show()
```
