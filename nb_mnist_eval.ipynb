```python
use_attn=True

mnist_cfg = draw_mnist.make_mnist_config(
    use_attn=use_attn,
    train_dir=("attn_" if use_attn else "") + "mnist_train_log")

tf.train.get_checkpoint_state(mnist_cfg.train_dir)
```

```python
gen_xs, gen_write_attn_infos = draw_mnist.eval_draw_model(mnist_cfg)
```

```python
batch_idx = 26
for i, gen_x in enumerate(gen_xs):
    x = gen_x[batch_idx,:,:]
    plt.imshow(x)
    plt.set_cmap("gray")
    if mnist_cfg.use_attn and i > 0:
        write_attn_info = [
            p[batch_idx] for p in gen_write_attn_infos[i - 1]
        ]
        gx, gy, sigma2, delta, gamma = write_attn_info
        w = delta * (mnist_cfg.write_width - 1.0)
        h = delta * (mnist_cfg.write_height - 1.0)
        print "debug: gx=%f, gy=%f, w=%f, h=%f, \nsigma2=%f, delta=%f, gamma=%f"%(
            gx, gy, w, h, sigma2, delta, gamma)
        g = (gx - w / 2.0, gy - h / 2.0)
        ax = plt.gca()
        ax.add_patch(patches.Rectangle(g, w, h, fill=False, edgecolor="r"))
    # plt.axis("off");    
    plt.show()
```
